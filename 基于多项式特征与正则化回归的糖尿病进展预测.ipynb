```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, learning_curve, KFold
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# 
from matplotlib.font_manager import FontProperties
myfont = FontProperties(fname=r'C:\Windows\Fonts\simhei.ttf', size=14)
sns.set(font=myfont.get_name())

# -
feature_name_map = {
    'age': '',
    'sex': '',
    'bmi': '',
    'bp': '',
    's1': '',
    's2': '',
    's3': '',
    's4': '',
    's5': '',
    's6': ''
}

def poly_feature_name_to_cn(name):
    for en, cn in feature_name_map.items():
        name = name.replace(en, cn)
    return name

# 1. 
def load_and_preprocess_data():
    diabetes = load_diabetes()
    X, y = diabetes.data, diabetes.target
    feature_names = diabetes.feature_names
    feature_names_cn = [feature_name_map[n] for n in feature_names]
    return X, y, feature_names, feature_names_cn

def polynomial_feature_expansion(X, feature_names, degree=2):
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly.fit_transform(X)
    feature_names_poly = poly.get_feature_names_out(feature_names)
    feature_names_poly_cn = [poly_feature_name_to_cn(n) for n in feature_names_poly]
    return X_poly, feature_names_poly, feature_names_poly_cn

def standardize_data(X):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    return X_scaled, scaler

# 2. 
def explore_data(X, y, feature_names_cn):
    # 
    corr = np.corrcoef(X.T)
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, xticklabels=feature_names_cn, yticklabels=feature_names_cn, cmap='coolwarm', annot=False)
    plt.title('')
    plt.tight_layout()
    plt.show()
    # 
    plt.figure(figsize=(8, 4))
    plt.hist(y, bins=30, color='skyblue')
    plt.title('')
    plt.xlabel('')
    plt.tight_layout()
    plt.show()

# 3. 
def perform_cross_validation(X, y, models, cv_folds=5):
    """
    
    """
    print(f"\n{'='*60}")
    print(f"{cv_folds}")
    print(f"{'='*60}")
    
    cv_results = {}
    kf = KFold(n_splits=cv_folds, shuffle=True, random_state=42)
    
    for name, model in models.items():
        # MSE
        mse_scores = -cross_val_score(model, X, y, cv=kf, 
                                    scoring='neg_mean_squared_error')
        # R
        r2_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
        
        cv_results[name] = {
            'mse_scores': mse_scores,
            'r2_scores': r2_scores,
            'mse_mean': mse_scores.mean(),
            'mse_std': mse_scores.std(),
            'r2_mean': r2_scores.mean(),
            'r2_std': r2_scores.std()
        }
        
        print(f"\n{name}:")
        print(f"  MSE = {mse_scores.mean():.2f}  {mse_scores.std():.2f}")
        print(f"  R  = {r2_scores.mean():.4f}  {r2_scores.std():.4f}")
        print(f"  MSE: {[f'{score:.2f}' for score in mse_scores]}")
        print(f"  R:  {[f'{score:.4f}' for score in r2_scores]}")
    
    return cv_results

# 4. 
def visualize_cv_results(cv_results):
    """
    
    """
    models = list(cv_results.keys())
    
    # 
    mse_means = [cv_results[m]['mse_mean'] for m in models]
    mse_stds = [cv_results[m]['mse_std'] for m in models]
    r2_means = [cv_results[m]['r2_mean'] for m in models]
    r2_stds = [cv_results[m]['r2_std'] for m in models]
    
    # 
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # MSE
    x = np.arange(len(models))
    axes[0, 0].bar(x, mse_means, yerr=mse_stds, capsize=5, color='salmon', alpha=0.7)
    axes[0, 0].set_xticks(x)
    axes[0, 0].set_xticklabels(models, rotation=45)
    axes[0, 0].set_ylabel('MSE')
    axes[0, 0].set_title('MSE')
    axes[0, 0].grid(True, alpha=0.3)
    
    # R
    axes[0, 1].bar(x, r2_means, yerr=r2_stds, capsize=5, color='skyblue', alpha=0.7)
    axes[0, 1].set_xticks(x)
    axes[0, 1].set_xticklabels(models, rotation=45)
    axes[0, 1].set_ylabel('R')
    axes[0, 1].set_title('R')
    axes[0, 1].grid(True, alpha=0.3)
    
    # MSE
    mse_data = [cv_results[m]['mse_scores'] for m in models]
    axes[1, 0].boxplot(mse_data, labels=models)
    axes[1, 0].set_xticklabels(models, rotation=45)
    axes[1, 0].set_ylabel('MSE')
    axes[1, 0].set_title('MSE')
    axes[1, 0].grid(True, alpha=0.3)
    
    # R
    r2_data = [cv_results[m]['r2_scores'] for m in models]
    axes[1, 1].boxplot(r2_data, labels=models)
    axes[1, 1].set_xticklabels(models, rotation=45)
    axes[1, 1].set_ylabel('R')
    axes[1, 1].set_title('R')
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# 5. 
def train_and_evaluate_models(X_train, X_test, y_train, y_test):
    models = {
        '': LinearRegression(),
        '': Ridge(),
        'Lasso': Lasso(max_iter=3000),
        '': ElasticNet(max_iter=3000)
    }
    results = {}
    predictions = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        predictions[name] = y_pred
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        results[name] = {'MSE': mse, 'R2': r2}
    return results, predictions

def plot_predictions(y_test, predictions):
    plt.figure(figsize=(12, 6))
    plt.plot(y_test, label='', color='black')
    for name, y_pred in predictions.items():
        plt.plot(y_pred, '--', label=name)
    plt.legend()
    plt.title('')
    plt.tight_layout()
    plt.show()

# 6. 
def plot_scatter_comparison(y_test, predictions):
    plt.figure(figsize=(14, 10))
    model_names = list(predictions.keys())
    for i, name in enumerate(model_names):
        plt.subplot(2, 2, i+1)
        plt.scatter(y_test, predictions[name], alpha=0.6)
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
        plt.xlabel('')
        plt.ylabel('')
        plt.title(f'{name}  vs ')
        plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# 7. 
def hyperparameter_tuning_and_plot(X, y, cv=5):
    best_params = {}

    # Lasso
    lasso_alphas = np.logspace(-3, 1, 50)
    lasso = Lasso(max_iter=3000)
    grid_lasso = GridSearchCV(lasso, {'alpha': lasso_alphas}, cv=cv, scoring='neg_mean_squared_error')
    grid_lasso.fit(X, y)
    lasso_scores = -grid_lasso.cv_results_['mean_test_score']

    plt.figure(figsize=(8, 4))
    plt.semilogx(lasso_alphas, lasso_scores, 'b-', label='LassoMSE')
    plt.axvline(grid_lasso.best_params_['alpha'], color='r', linestyle='--', label='alpha')
    plt.xlabel('alpha')
    plt.ylabel('MSE')
    plt.title('Lasso - ')
    plt.legend()
    plt.tight_layout()
    plt.show()
    best_params['Lasso'] = grid_lasso.best_params_

    # Ridge
    ridge_alphas = np.logspace(-3, 3, 50)
    ridge = Ridge()
    grid_ridge = GridSearchCV(ridge, {'alpha': ridge_alphas}, cv=cv, scoring='neg_mean_squared_error')
    grid_ridge.fit(X, y)
    ridge_scores = -grid_ridge.cv_results_['mean_test_score']

    plt.figure(figsize=(8, 4))
    plt.semilogx(ridge_alphas, ridge_scores, 'g-', label='RidgeMSE')
    plt.axvline(grid_ridge.best_params_['alpha'], color='r', linestyle='--', label='alpha')
    plt.xlabel('alpha')
    plt.ylabel('MSE')
    plt.title(' - ')
    plt.legend()
    plt.tight_layout()
    plt.show()
    best_params[''] = grid_ridge.best_params_

    # ElasticNet
    enet_alphas = np.logspace(-3, 1, 10)
    enet_l1_ratios = np.linspace(0.1, 0.9, 9)
    enet = ElasticNet(max_iter=3000)
    grid_enet = GridSearchCV(enet, {'alpha': enet_alphas, 'l1_ratio': enet_l1_ratios}, cv=cv, scoring='neg_mean_squared_error')
    grid_enet.fit(X, y)
    enet_results = pd.DataFrame(grid_enet.cv_results_)

    # ElasticNet
    pivot = enet_results.pivot(index='param_l1_ratio', columns='param_alpha', values='mean_test_score')
    plt.figure(figsize=(10, 6))
    sns.heatmap(-pivot, annot=True, fmt=".1f", cmap='YlOrRd')
    plt.title(' - MSE')
    plt.xlabel('alpha')
    plt.ylabel('l1_ratio')
    plt.tight_layout()
    plt.show()
    best_params[''] = grid_enet.best_params_

    print("\n", best_params)
    return best_params

# 8. 
def plot_learning_curves(X, y, models, cv=5):
    for name, model in models.items():
        train_sizes, train_scores, test_scores = learning_curve(
            model, X, y, cv=cv, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))
        train_mean = np.mean(train_scores, axis=1)
        test_mean = np.mean(test_scores, axis=1)
        plt.figure(figsize=(8, 5))
        plt.plot(train_sizes, train_mean, label='')
        plt.plot(train_sizes, test_mean, label='')
        plt.title(f'{name} ')
        plt.xlabel('')
        plt.ylabel('')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# 9. 
def plot_feature_importance(model, feature_names_poly_cn, top_n=20):
    coef = model.coef_
    top_idx = np.argsort(np.abs(coef))[-top_n:]
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(top_idx)), np.abs(coef[top_idx]))
    plt.yticks(range(len(top_idx)), [feature_names_poly_cn[i] for i in top_idx])
    plt.title(f'LassoTop{top_n}')
    plt.tight_layout()
    plt.show()

# 10. 
def main():
    # 1. 
    X, y, feature_names, feature_names_cn = load_and_preprocess_data()
    X_poly, feature_names_poly, feature_names_poly_cn = polynomial_feature_expansion(X, feature_names, degree=2)
    X_scaled, scaler = standardize_data(X_poly)
    
    # 2. 
    explore_data(X, y, feature_names_cn)
    
    # 3. 
    models = {
        '': LinearRegression(),
        '': Ridge(),
        'Lasso': Lasso(max_iter=3000),
        '': ElasticNet(max_iter=3000)
    }
    
    # 4. 
    cv_folds = 5  # 
    cv_results = perform_cross_validation(X_scaled, y, models, cv_folds)
    
    # 5. 
    visualize_cv_results(cv_results)
    
    # 6. 
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    print(f"\n{'='*60}")
    print("")
    print(f"{'='*60}")
    
    results, predictions = train_and_evaluate_models(X_train, X_test, y_train, y_test)
    for name, metrics in results.items():
        print(f"\n{name}:")
        print(f"  MSE = {metrics['MSE']:.2f}")
        print(f"  R  = {metrics['R2']:.4f}")
    
    plot_predictions(y_test, predictions)
    
    # 7. 
    plot_scatter_comparison(y_test, predictions)
    
    # 8. 
    best_params = hyperparameter_tuning_and_plot(X_scaled, y, cv=cv_folds)
    
    # 9. 
    optimized_models = {
        '': LinearRegression(),
        'Lasso': Lasso(alpha=best_params['Lasso']['alpha'], max_iter=3000),
        '': Ridge(alpha=best_params['']['alpha']),
        '': ElasticNet(alpha=best_params['']['alpha'],
                                  l1_ratio=best_params['']['l1_ratio'],
                                  max_iter=3000)
    }
    
    # 10. 
    print(f"\n{'='*60}")
    print("")
    print(f"{'='*60}")
    cv_results_optimized = perform_cross_validation(X_scaled, y, optimized_models, cv_folds)
    
    # 11. 
    plot_learning_curves(X_scaled, y, optimized_models, cv=cv_folds)
    
    # 12. 
    lasso = optimized_models['Lasso']
    lasso.fit(X_scaled, y)
    plot_feature_importance(lasso, feature_names_poly_cn, top_n=20)

if __name__ == '__main__':
    main()
```
